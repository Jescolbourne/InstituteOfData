# Hypothesis Testing

```python
#Define Hypothesis

# Null Hypothesis Dictionary
null_hypothesis = {'H0': "Web visits do not contribute to online sales"
    , 'H1': "Web visits contribute to online sales"}
```

```python
def print_sig(p_value, alpha):
    '''
    Inputs:
    p_value: Calculated p_value
    alpha: Confidence level
    '''
    if p_value < alpha:
        print("We reject our null hypothesis.")
        print(null_hypothesis['H1'])
    elif p_value > alpha:
        print("We fail to reject our null hypothesis.")
        print(null_hypothesis['H0'])
    else:
        print("Our test is inconclusive.")
```

```python
#Collect Data
```

```python
cp['numwebvisitsmonth'].describe()
```

```python
N = 1000
more = cp[cp['numwebvisitsmonth'] > 5]['numwebpurchases'].sample(1000, random_state=42)
less = cp[cp['numwebvisitsmonth'] <=5]['numwebpurchases'].sample(1000, random_state=42)
```

```python
#Set alpha
alpha = 0.05
```

```python
#Calculate variance to get the standard deviation
```

```python
more_var = more.var()
print(more_var)
less_var = less.var()
print(less_var)
```

```python
#calculate Standard Deviation
std = np.sqrt((more_var + less_var)/2)
print('std dev: ', std)
```

```python
#calculate test statistic
t = (more.mean() - less.mean()) / (std*np.sqrt(2/N))
print('t: ', t) #T-statistic
```

Find the p-value and Compare witht the t-value

```python
#calculate degree of freedom
df = 2*N-2
df
```

```python
#p-value after comparison with the t 
if (t > 0):
    p = 1 - stats.t.cdf(t, df = df) 
else:
    p = stats.t.cdf(t, df = df)
    
print("t = " + str(t))
print("p = " + str(2*p))
```

Interpret the results

```python
print("t = " + str(t))
print("p = " + str(p*2))
print_sig(p*2, alpha)
```
